```python
from resemblyzer import preprocess_wav, VoiceEncoder
import glob
import re
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data_directory = "./audio-samples/**/*.flac"
```

```python
filenames = glob.glob("./audio-samples/**/*.flac", recursive=True)
```

```python
encoder = VoiceEncoder()
```

```python
speaker_embeds = {}
for filename in tqdm(filenames):
    speaker = re.findall(r'\\([0-9]*)\\',filename)[0]
    waveform = preprocess_wav(filename)
    embed = encoder.embed_utterance(waveform)
    speaker_embeds[speaker] = speaker_embeds.get(speaker,[]) + [embed]
```

```python
# Lables follow the pattern of "speakerid-recordingnum"
lables = []
for key in speaker_embeds:
    for i in range(len(speaker_embeds[key])):
        lables += [key+"-"+str(i)]
```

```python
embed_matrix = np.concatenate(list(speaker_embeds.values()))
```

```python
sims = np.inner(embed_matrix, embed_matrix)
```

```python
sns.heatmap(sims,xticklabels=lables,yticklabels=lables)
```
